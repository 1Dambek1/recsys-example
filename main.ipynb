```python
import pandas as pd
```

```python
df =pd.read_csv("make_hard_blend.csv")
```

```python
def soft_voting(x):
 vars = {}
 for i in range(1,6):
  try:
   vars[x[f"model_{i}"]] += 1
  except:
   vars[x[f"model_{i}"]] = 1
 return max(vars, key=vars.get)
```

```python
df["target_class"] = df.apply(soft_voting, axis=1)
```

```python
df[["car_id","target_class"]].set_index("car_id").to_csv("submission.csv")
```

```python


cat = pd.read_csv("votes/catboost_preds.csv")
lgbm = pd.read_csv("votes/lgbm_preds.csv")
xgb = pd.read_csv("votes/xgb_preds.csv")
```

```python
ans = cat+lgbm+xgb
```

```python
def voting(x:pd.Series):
 idx = x.index[x==max(x)]
 return "_".join(idx.item().split("_")[0:2])
```

```python
ans["soft_preds"] = ans.apply(voting, axis=1)
```

```python
ans[["soft_preds"]].to_csv("submission.csv")
```

```python
df= pd.read_csv("meta_feautres_data.csv")

df
```

```python
df = pd.pivot_table(df,
               index = "index",
               columns="model",
               values="target",
               ).reset_index()
```

```python
df.columns = ["index","target_catboost", "target_lgbm", "target_xgb"]
```

```python
for i in ["catboost", "lgbm", "xgb"]:
 for j in ["catboost", "lgbm", "xgb"]:
  df[f"{i}_dif_{j}"] = df[f"target_{i}"]-df[f"target_{j}"]
  df[f"{i}_div_{j}"] = df[f"target_{i}"]/df[f"target_{j}"]
```

```python
df.set_index("index").to_csv("submission.csv")
```
